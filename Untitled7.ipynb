```python
import nltk
#nltk.download("stopwords")
#--------#

from nltk.corpus import stopwords
from pymystem3 import Mystem
from string import punctuation
import io
mystem = Mystem()
russian_stopwords = stopwords.words("russian")
def preprocess_text_opt(text):
    text = text.lower().split(" ")
    tokens = [token for token in text if token not in russian_stopwords\
              and token != " " \
              and token.strip() not in punctuation]
    if len(tokens) < 100:
        return tokens
    else:
        return tokens[0:100]
def fun(a, b):
    count = 0
    for r in range(a, b):
        lemmer = {}
        index = {}
        print(r)
        cur = ""
        with io.open("./files/выкачка "+str(r)+".txt", encoding='utf-8') as file:
            for line in file:
                cur =cur +  " " + line
        cur.replace("\n", '')
        cur = cur.lower()
        words = preprocess_text_opt(cur)
        print(len(words))
        for d in range(0, len(words)):
            i = words[d]
            if d%25==0:
                print(d/len(words))
            d = mystem.lemmatize(i)[0]
            value = lemmer.get(d)
            if d in lemmer.keys():
                if i not in lemmer[d]:
                    lemmer[d].append(i)
            else:
                lemmer[d] = [i]
            if d in index.keys():
                if r not in index[d]:
                    index[d].append(r)
            else:
                index[d] = [r]
        with open("./work/lemmer "+str(r)+".txt", 'w', encoding='utf-8') as file:
            for i in lemmer.keys():
                file.write(i+" ")
                for w in lemmer[i]:
                    file.write(w + " ")
                file.write("\n")
        with open("./work/index "+str(r)+".txt", 'w', encoding='utf-8') as file:
            for i in index.keys():
                file.write(i+" ")
                for w in index[i]:
                    file.write(str(w) + " ")
                file.write("\n")
```
