```python
import nltk
#nltk.download("stopwords")
#--------#

from nltk.corpus import stopwords
from pymystem3 import Mystem
from string import punctuation
import io
mystem = Mystem()
russian_stopwords = stopwords.words("russian")
import pymorphy2
def preprocess_text_opt(text):
    text = text.lower().split(" ")
    tokens = [token for token in text if token not in russian_stopwords\
              and token != " " \
              and token.strip() not in punctuation]
    return tokens
lemmer = {}
index = {}
w = []
def fun(a, b):
    morph = pymorphy2.MorphAnalyzer()
    morph = pymorphy2.MorphAnalyzer(lang='ru')
    count = 0
    for r in range(a, b):
        print(r)
        cur = ""
        with io.open("./files/выкачка "+str(r)+".txt", encoding='utf-8') as file:
            for line in file:
                cur =cur +  " " + line
        cur = cur.replace("\n", '')
        cur = cur.replace("}", "")
        cur = cur.replace("{", "")
        cur = cur.replace("-", "")
        cur = cur.replace("—", "")
        cur = cur.replace("\n", "")
        cur = cur.replace("(", "")
        cur = cur.replace(")", "")
        cur = cur.replace(".", "")
        cur = cur.replace(",", "")
        cur = cur.replace(":", "")
        cur = cur.replace("[", "")
        cur = cur.replace("]", "")
        cur = cur.replace("'", "")
        cur = cur.replace("፡", '')
        cur = cur.replace(";", "")
        cur = cur.replace("?", "")
        cur = cur.replace('"', "")
        cur = cur.replace("/", "")
        cur = cur.replace("_", "")
        cur = cur.replace('=', "")
        cur = cur.replace('&', "")
        cur = cur.replace('…', "")
        cur = cur.lower()
        words = preprocess_text_opt(cur)
        txt = ""
        for i in words:
            txt = txt + i + " "
        lems = mystem.lemmatize(txt)
        words = txt.split(" ")
        lems = [i for i in lems if i != " "]
        for b in range(0, len(words)):
            i = words[b]
            w.append(i)
            d = morph.parse(i)[0].normal_form
            if d != " ": 
                value = lemmer.get(d)
                if d in lemmer.keys():
                    if i not in lemmer[d]:
                        lemmer[d].append(i)
                else:
                    lemmer[d] = [i]
                if d in index.keys():
                    if r not in index[d]:
                        index[d].append(r)
                else:
                    index[d] = [r]
                    
       
```
